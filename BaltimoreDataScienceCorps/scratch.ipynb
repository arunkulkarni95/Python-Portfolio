```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('Zip_MHINC_Pop.csv')


df = df.drop(columns=['Unnamed: 0'])
print(df.head())
```

```python
incMean = np.mean(df['TotalCases'])
incStd = np.std(df['TotalCases'])

samples = np.random.normal(incMean,incStd,size=1000000)
```

```python
def ecdf(df):
    n = len(df)
    x = np.sort(df)
    y = np.arange(1, n + 1) /n
    return x, y
```

```python
x_theor, y_theor = ecdf(samples)
x, y = ecdf(df['TotalCases'])

_ = plt.figure(figsize=(12,6.75))
_ = plt.style.use('ggplot')
_ = plt.plot(x_theor,y_theor)
_ = plt.plot(x,y,marker='.',linestyle='none')
_ = plt.xlabel('Total COVID-19 Cases')
_ = plt.ylabel('CDF')
_ = plt.title('CDF of Baltimore COVID-19 vs. Normal Distribution')
plt.show()
```
